```python
num1 = float(input("첫 번째 숫자: "))
operator = input("연산자(+, -, *, /): ")
num2 = float(input("두 번째 숫자: "))

if operator == '+':
  result = num1 + num2
elif operator == '-':
  result = num1 - num2
elif operator == '*':
  result  = num1 * num2
elif operator == '/':
  if num2 != 0:
    result = num1 / num2
  else:
    print('0으로 나눌 수 없습니다')
    result = None
else :
  print('잘못된 연산자 입니다.')
  result = None

if result is not None:
 print(f"결과: {num1} {operator} {num2} = {result}")
```
